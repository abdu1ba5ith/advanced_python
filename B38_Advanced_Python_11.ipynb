```python
x = [1,2,3]
y = [1,4,9]
```

```python
x = [1,2,3]
y = [1,4,9]



plt.plot(x,y)
plt.xlabel('Sales')
plt.ylabel('Revenue')
plt.show()
```

```python
x = [1,2,3]
y = [1,4,9]



plt.plot(x,y)
plt.xlabel('Sales')
plt.ylabel('Revenue')
plt.title('Performance Calulator of Sales')
plt.show()
```

```python
x = [1,2,3]
y = [1,4,9]
z = [10,5,0]


plt.plot(x,y)  # this is x,y
plt.plot(x,z)  # This is x,z 
plt.xlabel('Sales')
plt.ylabel('Revenue')
plt.title('Performance Calulator of Sales')
plt.show()

```

```python
x = [1,2,3]
y = [1,4,9]
z = [10,5,0]


plt.plot(x,y)  # this is x,y  ==> Data from credit customers
plt.plot(x,z)  # This is x,z ==> Data from Corporate Customers
plt.xlabel('Sales')
plt.ylabel('Revenue')
plt.title('Performance Calulator of Sales')
plt.legend(['this is x and y','this is x and z'])
plt.show()
```
